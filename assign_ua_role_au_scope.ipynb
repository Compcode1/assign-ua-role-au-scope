**Assign User Administrator Role at AU Scope**

**Scenario**
To support departmental autonomy, we need to delegate user management permissions within a specific division of the organization—Marketing—without granting global directory rights. This is done by assigning the built-in User Administrator role scoped to the Marketing Administrative Unit (AU). This ensures that the delegated admin can manage users only within their department, preserving governance and containment boundaries.

**Step-by-Step Action Flow (Simulated)**

Navigate to Microsoft Entra Admin Center → Identity → Administrative Units

Select Marketing AU → Roles and administrators

Click User Administrator

Click Add assignment

Select user: aurelia.lee@company.com

Confirm scope: Marketing (not directory-wide)

Click Assign

Result: User aurelia.lee@company.com is now assigned the User Administrator role scoped only to the Marketing AU.

**Entra Control Stack Mapping**

**Layer 1 – Authority Definition**
✅ Touched
The action is performed by a Global Administrator or a Privileged Role Administrator who has the authority to assign roles. Role assignment audit logs are generated.

**Layer 2 – Scope Boundaries**
✅ Primary control layer
This project directly engages scope limitation by assigning the User Administrator role only within the Marketing AU. This enforces strong boundary controls and is central to least-privilege governance.

**Layer 3 – Test Identity Validation**
✅ Essential follow-up
After assignment, validate effective permissions:

Confirm that aurelia.lee can reset passwords for Marketing users.

Attempt to access users outside the AU; access should be denied.

Optionally use **Microsoft Graph** or **Entra Portal** to simulate and verify role behavior.

**Layer 4 – External Entry Controls**
❌ Not affected
This action applies to internal user role scoping and does not involve external or B2B identities.

**Layer 5 – Privilege Channels**
✅ Moderate control activation
A privileged role (User Administrator) is granted in a scoped and minimized way. The privilege channel is opened only within a defined boundary, reducing global risk.

**Layer 6 – Device Trust Enforcement**
❌ Not affected
This project does not intersect with device compliance or enforcement.

**Layer 7 – Continuous Verification**
✅ Post-assignment hygiene
Monitor sign-ins and role activity for aurelia.lee
Add this assignment to periodic access review cycles
Ensure the scope remains limited to the intended AU and is not elevated over time

**Observations and Follow-Up**
Scoped role assignment was successful; access is limited to the Marketing AU.
Attempted access to users in other departments was denied, confirming correct enforcement.
Audit logs reflect the role assignment and any subsequent actions by aurelia.lee.
Recommend this AU-based delegation model be used broadly to segment admin rights by department or geography.
